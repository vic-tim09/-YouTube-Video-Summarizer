```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# extract the video ID from the YouTube URL

from urllib.parse import urlparse , parse_qs

url = input("Paste YouTube URL: ")

parsed = urlparse(url)

if "youtu.be" in parsed.netloc:
    video_id = parsed.path.lstrip("/")
else: 
    video_id = parse_qs(parsed.query)["v"][0]

print(video_id)
```

```python
# fetch the transcript for the video
transcript = YouTubeTranscriptApi().fetch(video_id)

print(transcript[:5])
```

```python
# put the transcript into a single string
text = " ".join(
    [item.text for item in transcript]
)
print("Total Characters:", len(text))
```

```python
# Save the transcript to a text file
with open("transcript.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("Transcript Saved")
```

```python
# Initialize Groq with your API key
from dotenv import load_dotenv
import os
from groq import Groq

load_dotenv()

api_key = os.getenv("GROQ_API_KEY")

client = Groq(api_key=api_key)
```

```python
with open("transcript.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(len(text))
```

```python
# Step 1: Add proper punctuation and capitalization to the transcript
response1 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{
        "role": "user",
        "content": f"Add proper punctuation and capitalization to this transcript. Return only the corrected text:\n\n{text}"
    }]
)
punctuated_text = response1.choices[0].message.content

# Step 2: Summarize the punctuated transcript
response2 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{
        "role": "user",
        "content": f"""
Summarize this YouTube transcript.

Give:
1. Main Topic
2. Key Points
3. Important Insights
4. Final Conclusion

Transcript:
{punctuated_text}
"""
    }]
)

summary = response2.choices[0].message.content
print(summary)
```
